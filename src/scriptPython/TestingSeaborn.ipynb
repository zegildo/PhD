```python
%matplotlib inline

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sns.set(style="whitegrid")

volNewsYear = pd.read_csv('/home/zegildo/workspace/PhD/src/scriptPython/volumeNoticias-Ano.csv')

# Draw a nested barplot to show survival for class and sex
g = sns.factorplot(x="ano", y="quantidade", hue="jornal", data=volNewsYear,
                   size=6, kind="bar", palette="muted",  aspect=2)
g.set_ylabels("Quantida de Noticipas Publicadas por Ano")
```

```python
volTotal = pd.read_csv('/home/zegildo/workspace/PhD/src/scriptPython/volumeTotalGranularidades.csv')

sns.set(style="darkgrid")
g = sns.FacetGrid(volTotal, row="Ano", col="Jornal", margin_titles=True)
bins = np.linspace(0, 230, 35)
g.map(plt.hist,"size", color="steelblue", bins=bins, lw=0)
```

```python
volMonths = pd.read_csv('/home/zegildo/workspace/PhD/src/scriptPython/volumeNoticias-Mes.csv')

g = sns.factorplot(x="mes", y="quantidade", hue="jornal", data=volMonths,
                   size=6, kind="point", palette="muted",  aspect=2)
g.set_ylabels("Quantida de Noticipas Publicadas por Meses")
```
